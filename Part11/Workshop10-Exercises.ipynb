## Exercise 2

Given the following mathematical functions:

$$f(x) = 5x$$

$$f(x) = 3x + 2$$

$$f(x) = -\frac{1}{2}x + 5$$

Create a graph that plots all three functions in the interval $[0, 7]$. Make sure to include a legend, axes labels, and a title.

````python
import math
import numpy
fx1=numpy.arange(0,7)
fy1=[5*x for x in fx1]
fy2=[3*x+2 for x in fx1]
fy3=[-(x/2)+5 for x in fx1]
plt.style.use("default")
plt.plot(fx1,fy1, label="f(x)1")
plt.plot(fx1,fy2, label="f(x)2")
plt.plot(fx1,fy3, label="f(x)3")
plt.legend()
plt.title("this is title")
plt.show()
#print(fy3)
````

## Exercise 3

Repeat the exercise above with a ```for loop```.

````python
import math
import numpy
fx=numpy.arange(0,7)
fy=[[5*x for x in fx],[3*x+2 for x in fx],[-(x/2)+5 for x in fx]]
print(fy)
for y in fy:
    plt.plot(fx,y)
plt.style.use("default")
plt.legend(["f(x)1","f(x)2","f(x)3"])
plt.title("this is title")
plt.xlabel("this x lable")
plt.ylabel("this y lable")
plt.show()
````

## Exercise 4

Given the following mathematical functions:

$$ f(x) = \sin(2x) $$
$$ f(x) = \cos(x) $$
$$ f(x) = 2\sin(x) + \cos(2x) $$

Create a graph that plots all three functions in the interval $[0, 4\pi]$. Make sure to include a legend, axes labels, and a title.

Import the ```sin()``` and ```cos()``` functions and the constant ```pi``` from the ```math``` module, and the ```linspace()``` function from the ```numpy``` library.

Use the ```help()``` function or ```fun_name?``` command to get the relevant documentation.

````python
from math import sin,cos,pi
from numpy import linspace
xline=linspace(0,4*pi,100)
yline1=[sin(2*x) for x in xline]
yline2=[cos(x) for x in xline]
yline3=[2*sin(x)+cos(2*x) for x in xline]
plt.style.use("default")
plt.plot(xline,yline1,label="fun1=sin(2x)")
plt.plot(xline,yline2,label="fun2=cos(x)")
plt.plot(xline,yline3,label="fun3=2sin(x)+cos(2x)")
plt.legend()
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title("this is Fun Test")
plt.show()
````

## Exercise 5

Repeat the exercise above with a ```for loop```.

````python
from math import sin,cos,pi
from numpy import linspace
xline=linspace(0,4*pi,100)
yline=[[sin(2*x) for x in xline],[cos(x) for x in xline],[2*sin(x)+cos(2*x) for x in xline]]
for y in yline:
    plt.plot(xline,y)
plt.style.use("default")
plt.legend(["fun1=sin(2x)","fun2=cos(x)","fun3=2sin(x)+cos(2x)"])
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title("this is Fun Test")
plt.show()
````
