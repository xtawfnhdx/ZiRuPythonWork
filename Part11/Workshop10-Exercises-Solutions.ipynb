## Exercise 2

Given the following mathematical functions:

$$f(x) = 5x$$

$$f(x) = 3x + 2$$

$$f(x) = -\frac{1}{2}x + 5$$

Create a graph that plots all three functions in the interval $[0, 7]$. Make sure to include a legend, axes labels, and a title.

````python
# Create a range of values from 0 to 7 with the range() function. The plt.plot() function can accept any "array-like" arguments that represent a sequence of values
x_range = range(8)

# Create associated y values according to the functions above. Use a list comprehension instead of a for loop to create the values.
y1 = [5*x for x in x_range]
y2 = [3*x + 2 for x in x_range]
y3 = [-x/2 + 5 for x in x_range]

# Plot the lines by subsequent calls to plt.plot()
plt.plot(x_range, y1)
plt.plot(x_range, y2)
plt.plot(x_range, y3)

# Add information
plt.legend(["5x", "3x + 2", "-1/2 x + 5"])
plt.xlabel("x")
plt.ylabel("y")
plt.title("Three linear functions")

plt.show()
````

## Exercise 3

Repeat the exercise above with a ```for loop```.

````python
# For each of the above mathematical functions, define a Python function that returns a list of values given a list as an argument
fun1 = lambda x_range: [5*x for x in x_range]
fun2 = lambda x_range: [3*x + 2 for x in x_range]
fun3 = lambda x_range: [-x/2 + 5 for x in x_range]

# Write a for loop that uses the lambda function to subsequently plot the correct line
for fun in [fun1, fun2, fun3]:
    plt.plot(x_range, fun(x_range))

# Add information
plt.legend(["5x", "3x + 2", "-1/2 x + 5"])
plt.xlabel("x")
plt.ylabel("y")
plt.title("Three linear functions")

plt.show()
````

## Exercise 4

Given the following mathematical functions:

$$ f(x) = \sin(2x) $$
$$ f(x) = \cos(x) $$
$$ f(x) = 2\sin(x) + \cos(2x) $$

Create a graph that plots all three functions in the interval $[0, 4\pi]$. Make sure to include a legend, axes labels, and a title.

Import the ```sin()``` and ```cos()``` functions and the constant ```pi``` from the ```math``` module, and the ```linspace()``` function from the ```numpy``` library.

Use the ```help()``` function or ```fun_name?``` command to get the relevant documentation.

````python
# Import the necessary functions from the maths and numpy libraries
from math import sin, cos, pi
from numpy import linspace

# Create a range of equally spaced x values using linspace()
x_values = linspace(0, 4*pi, 100)

# Create y values associated to the trigonometric functions above
y1 = [sin(2*x) for x in x_values]
y2 = [cos(x) for x in x_values]
y3 = [2*sin(x) + cos(2*x) for x in x_values]

# Plot the lines by subsequent calls to plt.plot()
plt.plot(x_values, y1)
plt.plot(x_values, y2)
plt.plot(x_values, y3)

# Add information
plt.legend(["sin(2x)", "cos(x)", "2sin(x) + cos(2x)"])
plt.xlabel("x")
plt.ylabel("y")
plt.title("Three trigonometric functions")

plt.show()
````

## Exercise 5

Repeat the exercise above with a ```for loop```.

````python
fun1 = lambda x_values: [sin(2*x) for x in x_values]
print(fun1)
````

````python
# For each of the above mathematical functions, define a Python function that returns a list of values given a list as an argument
fun1 = lambda x_values: [sin(2*x) for x in x_values]
fun2 = lambda x_values: [cos(x) for x in x_values]
fun3 = lambda x_values: [2*sin(x) + cos(2*x) for x in x_values]

# Write a for loop that uses the lambda function to subsequently plot the correct line
for fun in [fun1, fun2, fun3]:
    plt.plot(x_values, fun(x_values))
    
# Add information
plt.legend(["sin(2x)", "cos(x)", "2sin(x) + cos(2x)"])
plt.xlabel("x")
plt.ylabel("y")
plt.title("Three trigonometric functions")

plt.show()
````

## Exercise 7

The file ```health_data.csv``` contains a small dataset with two variables for a set of countries. For each country, the dataset reports

- The life expectancy at birth measured in years
- The health expenditure per capita

Import the dataset and plot a scatterplot where health expenditure is the predictor variable and life expectancy is the predicted variable.

To create a scatterplot, you can use the ```plt.scatter()``` function from Matplotlib. Use the ```help()``` function or read the [official documentation](https://matplotlib.org/3.3.3/api/_as_gen/matplotlib.pyplot.scatter.html#matplotlib-pyplot-scatter) to understand the required syntax.

````python
df_health = pd.read_csv("health_data.csv")
df_health
````

````python
# Create a scatter plot using the plt.scatter() function
# Since health expenditure is our predictor variable, we will show that on the horizontal axis
plt.scatter("Health expenditure", "Life expectancy", data=df_health)

# Add information
plt.xlabel("Health expenditure")
plt.ylabel("Life expectancy")
plt.title("Scatter plot")

plt.show()
````

As you can see the relationship does not seem to be linear. As we discussed in a previous workshop, it might be a good idea to transform the health expenditure variable using the natural logarithm.

````python
# Import the natural logarithm function from numpy
from numpy import log
````

````python
# Create an additional column to the data frame containing the log-trasformed variable
df_health["log health exp"] = log(df_health["Health expenditure"])
df_health.head()
````

````python
# Recreate the scatterplot with the log-transformed variable
plt.scatter("log health exp", "Life expectancy", data=df_health)

# Add information
plt.xlabel("log health exp")
plt.ylabel("Life expectancy")
plt.title("Scatter plot (log)")

plt.show()
````
