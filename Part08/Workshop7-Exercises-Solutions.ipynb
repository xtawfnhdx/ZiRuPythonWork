## Exercise 19
Given the following basket

| **Good**      | **Price**       | **Quantity** |
| :------------ | :------------:  | :----------: |
| Cola          | 1.25            | 5            |
| Water         | 0.75            | 7            |
| Chocolate     | 3               | 4            |

and the following lists

```python
prices     = [1.25, 0.75, 3]
quantities = [5, 7, 4]
```

Compute and display the total cost of the basket **using a for loop**.

````python
# Prices and quantities
prices     = [1.25, 0.75, 3]
quantities = [5, 7, 4]

# Approach 1: loop over prices and quantities using indexes from the range() function
# Create a variable cost of basket. The cost of the basket is now zero
cost_basket = 0

for i in range(len(prices)):
    # Increment the cost of the basket with the cost of the current good
    cost_basket += prices[i]*quantities[i]

print(f"The total cost of the basket is £{cost_basket}")

# Approach 2: loop over pairs of price and quantity with zip() function
cost_basket = 0

for price, quantity in zip(prices, quantities):
    # Increment the cost of the basket with the cost of the current good
    cost_basket += price*quantity

print(f"The total cost of the basket is £{cost_basket}")
````

## Exercise 24

Given the two lists below:

````python
years = list(range(1980,2000))
print(years)

inflation = [20.82, 18.79, 17.54, 15.10, 10.76, 9.17, 7.49, 6.01, 6.65, 6.20, 8.91, 7.58, 4.37, 3.89, 3.54, 4.93, 4.55, 2.60, 2.53, 1.62]
print(inflation)
````

Create a dictionary where years are the keys and inflation are the values. Print out all the years and correspondent inflation values using a for loop.

````python
# The aim is to construct a dictionary, that is a sequence of key: value pairs, from the existing lists
# Eventually, we will want to use the dict() function that creates a dictionary from another data type
help(dict)
````

````python
# As the help function shows (third usage), the dict() function allows to create a dictionary
# from an iterable containing pairs of (key, value), that is a list of tuples
# The zip() function can create that iterable object of (key, value) tuples
print(zip(years, inflation))
````

````python
# As this is a zip object, it is not directly printable. Let's convert it into a list. This is just for printing purposes
print(list(zip(years, inflation)))
````

````python
# We can now use the dict() function to create the dictionary. We will pass in the zip() object, as it is an iterable.
inflation_data = dict(zip(years, inflation))

# Check if dictionary is created correctly
print(inflation_data)
````

Notice that the keys of the ```inflation_data``` dictionary are ```integers```.

Up to know we worked with keys being strings. However, keys can be any **immutable** data type. Integers is an example.

Having keys as integers allows us to index the dictionary with numbers, and to manipulate those numbers to get new keys. Let's say that we want to get inflation for 1980 and inflation for 1982. We can do that as follows:

````python
print(inflation_data[1980])
print(inflation_data[1980+2])
````
